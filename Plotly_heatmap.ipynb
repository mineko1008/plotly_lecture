
# ヒートマップの描画


## 1.ライブラリのインポート

今回はPlotly figure_factoryを使ってヒートマップを描画していきます。


以下のセルを実行して、Plotly figure_factoryをインポートします。
"as ff"としているので、今後は"ff"で呼び出すことができます。





```python
import plotly.figure_factory as ff
```

## 2.データの準備

今回は下記の配列を使います。\
変数に代入しておきます。

```python
z = [[.1, .3, .5],
     [1.0, .8, .6],
     [.6, .4, .2]]
```

## 3.ヒートマップを描く
まずはベーシックなヒートマップを描いてみましょう！\
ff.create_annotated_heatmap(**使用するデータ**)

fig=ff.create_annotated_heatmap()がfigure_factory(ff)でヒートマップ(heatmap)を使いますよという意味です。\
一度変数に代入して、それをshowで表示しています。\
変数は今回は"fig"としていますが、何でも大丈夫です。figure=図を意味しているので良く使われる表現です。

値が小さいものは青に、大きくなるほど黄色に色がつけられています。




```python
fig = ff.create_annotated_heatmap(z)
fig.show()
```

**colorscale**で色を指定することが出来ます。\
xとｙに見出しを付けておきましょう。

```python
fig = ff.create_annotated_heatmap(z, 
                                  colorscale='Reds',
                                  x=['Team A', 'Team B', 'Team C'],
                                  y=['Game One', 'Game Two', 'Game Three'])
fig.show()
```

update_layoutで、細かなところを調整していきましょう。\
ここでは、大きさを調整し、タイトルを追加しています。\
また、y軸を並び替えて、軸ラベルを追加しました。

```python
fig = ff.create_annotated_heatmap(z, 
                                  colorscale='Reds',
                                  x=['Team A', 'Team B', 'Team C'],
                                  y=['Game One', 'Game Two', 'Game Three'])
fig.update_layout(height=400, 
                  width=600, 
                  title_text='試合結果',
                  yaxis=dict(autorange='reversed', title="GAME")
                  )
fig.show()
```

**annotation_text**で、ヒートマップの色はそのままに、表示されるテキストを指定することができます。\
ホバーすると値を確認することができます。

```python
import plotly.figure_factory as ff

z = [[.1, .3, .5],
     [1.0, .8, .6],
     [.6, .4, .2]]

x = ['Team A', 'Team B', 'Team C']
y = ['Game One', 'Game Two', 'Game Three']

z_text = [['Win', 'Lose', 'Win'],
          ['Lose', 'Lose', 'Win'],
          ['Win', 'Win', 'Lose']]

fig = ff.create_annotated_heatmap(z, 
                                  x=x, 
                                  y=y, 
                                  annotation_text=z_text, 
                                  colorscale='Reds')
fig.update_layout(height=400, 
                  width=600, 
                  title_text='試合結果',
                  yaxis=dict(autorange='reversed', title="GAME")
                  )
fig.show()
```
